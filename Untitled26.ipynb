```python
import torch
import torch.nn as nn
import torch.distributed as dist
from torch.nn.parallel import DistributedDataParallel as DDP
from torch.utils.data import Dataset, DataLoader, DistributedSampler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import json
from typing import Dict, Tuple, Optional
from dataclasses import dataclass
import time
from sklearn.metrics import roc_auc_score, roc_curve
import warnings
warnings.filterwarnings('ignore')


@dataclass
class Config:
    """Configuration for training pipeline"""
    # Data
    data_path: str = "/content/atlas-higgs-challenge-2014-v2.csv"

    # Model
    input_features: int = 31  # CSV has 31 numeric features (30 features + 1 label)
    hidden_dims: list = None
    dropout: float = 0.3

    # Training
    batch_size: int = 512
    epochs: int = 10
    learning_rate: float = 1e-3
    weight_decay: float = 1e-4

    # Distributed
    world_size: int = 1  # Number of GPUs
    backend: str = "nccl"  # Use NCCL for multi-GPU

    # Output
    model_path: str = "./models/higgs_classifier.pt"
    results_path: str = "./results/"

    def __post_init__(self):
        if self.hidden_dims is None:
            self.hidden_dims = [256, 128, 64, 32]
        Path(self.results_path).mkdir(exist_ok=True, parents=True)


class ATLASDataset(Dataset):
    """
    Dataset loader for ATLAS data from CSV format

    Loads atlas-higgs-challenge-2014-v2.csv
    """

    def __init__(self, csv_file: str, transform=None):
        """
        Args:
            csv_file: Path to CSV file with features and labels
            transform: Optional data transformation
        """
        self.transform = transform

        # Load CSV data
        try:
            print(f"Loading ATLAS data from CSV: {csv_file}")
            self.data = pd.read_csv(csv_file)
            print(f"✓ Loaded {len(self.data)} events")
            print(f"  Columns: {list(self.data.columns)}")

            # Detect label column
            if 'Label' in self.data.columns:
                self.label_col = 'Label'
            else:
                self.label_col = self.data.columns[0]

            # Map string labels to numeric
            if self.data[self.label_col].dtype == object:
                self.label_map = {'s': 1, 'b': 0}
                self.data[self.label_col] = self.data[self.label_col].map(self.label_map)
                # Fill any NaN values introduced by mapping with 0 (background) before converting to int
                self.data[self.label_col] = self.data[self.label_col].fillna(0)

            self.features = self.data.drop(columns={self.label_col, 'Weight', 'weight', 'EventId', 'Id', 'id'} & set(self.data.columns))
            self.features = self.features.select_dtypes(include=[np.number]).fillna(0)

            self.labels = self.data[self.label_col].astype(int)

        except Exception as e:
            print(f"⚠ Error loading CSV: {e}")
            raise

    def _load_csv_file(self, filepath: str) -> pd.DataFrame:
        """Load CSV file from path"""
        return pd.read_csv(filepath)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        # Handle both direct indexing and subset indexing from random_split
        if hasattr(self, 'indices') and self.indices is not None:
            idx = self.indices[idx]

        try:
            feature_vals = self.features.iloc[idx].values
        except (IndexError, KeyError):
            # Fallback: convert to numpy array for direct indexing
            feature_vals = self.features.values[idx]

        features = torch.tensor(feature_vals, dtype=torch.float32)
        label = torch.tensor(float(self.labels.iloc[idx]), dtype=torch.float32)

        if self.transform:
            features = self.transform(features)

        return features, label


class HiggsClassifier(nn.Module):
    """
    Deep Neural Network for Higgs Boson Detection
    Architecture based on successful approaches in ATLAS/CMS papers
    """

    def __init__(self, input_dim: int, hidden_dims: list, dropout: float = 0.3):
        super().__init__()

        layers = []
        prev_dim = input_dim

        for hidden_dim in hidden_dims:
            layers.extend([
                nn.Linear(prev_dim, hidden_dim),
                nn.BatchNorm1d(hidden_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            ])
            prev_dim = hidden_dim

        # Output layer
        layers.append(nn.Linear(prev_dim, 1))
        layers.append(nn.Sigmoid())

        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x).squeeze()


class DistributedTrainer:
    """
    Production trainer with multi-GPU support using PyTorch DDP
    """

    def __init__(self, config: Config, rank: int = 0):
        self.config = config
        self.rank = rank
        self.device = torch.device(f"cuda:{rank}" if torch.cuda.is_available() else "cpu")

        # Setup distributed if multi-GPU
        if config.world_size > 1:
            self._setup_distributed()

        # Build model
        self.model = HiggsClassifier(
            input_dim=config.input_features,
            hidden_dims=config.hidden_dims,
            dropout=config.dropout
        ).to(self.device)

        if config.world_size > 1:
            self.model = DDP(self.model, device_ids=[rank])

        # Loss and optimizer
        self.criterion = nn.BCELoss()
        self.optimizer = torch.optim.AdamW(
            self.model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay
        )
        self.scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer, mode='min', patience=5, factor=0.5
        )

        # Metrics
        self.history = {
            'train_loss': [], 'val_loss': [],
            'train_auc': [], 'val_auc': []
        }

    def _setup_distributed(self):
        """Initialize distributed training"""
        dist.init_process_group(
            backend=self.config.backend,
            init_method='env://',
            world_size=self.config.world_size,
            rank=self.rank
        )

    def train_epoch(self, train_loader):
        """Train for one epoch"""
        self.model.train()
        total_loss = 0
        all_preds, all_labels = [], []

        for features, labels in train_loader:
            features = features.to(self.device)
            labels = labels.to(self.device)

            # Forward
            self.optimizer.zero_grad()
            predictions = self.model(features)
            loss = self.criterion(predictions, labels)

            # Backward
            loss.backward()
            self.optimizer.step()

            total_loss += loss.item()
            all_preds.extend(predictions.detach().cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

        avg_loss = total_loss / len(train_loader)
        auc = roc_auc_score(all_labels, all_preds)

        return avg_loss, auc

    def validate(self, val_loader):
        """Validate model"""
        self.model.eval()
        total_loss = 0
        all_preds, all_labels = [], []

        with torch.no_grad():
            for features, labels in val_loader:
                features = features.to(self.device)
                labels = labels.to(self.device)

                predictions = self.model(features)
                loss = self.criterion(predictions, labels)

                total_loss += loss.item()
                all_preds.extend(predictions.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        avg_loss = total_loss / len(val_loader)
        auc = roc_auc_score(all_labels, all_preds)

        return avg_loss, auc, all_preds, all_labels

    def fit(self, train_loader, val_loader):
        """Full training loop"""
        print(f"\n{'='*70}")
        print(f"Training on device: {self.device}")
        print(f"{'='*70}\n")

        best_val_auc = 0

        for epoch in range(self.config.epochs):
            start_time = time.time()

            # Train
            train_loss, train_auc = self.train_epoch(train_loader)

            # Validate
            val_loss, val_auc, _, _ = self.validate(val_loader)

            # Learning rate scheduling
            self.scheduler.step(val_loss)

            # Save metrics
            self.history['train_loss'].append(train_loss)
            self.history['val_loss'].append(val_loss)
            self.history['train_auc'].append(train_auc)
            self.history['val_auc'].append(val_auc)

            epoch_time = time.time() - start_time

            # Print progress
            if self.rank == 0:
                print(f"Epoch {epoch+1}/{self.config.epochs} ({epoch_time:.1f}s)")
                print(f"  Train Loss: {train_loss:.4f} | AUC: {train_auc:.4f}")
                print(f"  Val Loss: {val_loss:.4f} | AUC: {val_auc:.4f}")
                print(f"  LR: {self.optimizer.param_groups[0]['lr']:.6f}\n")

            # Save best model
            if val_auc > best_val_auc and self.rank == 0:
                best_val_auc = val_auc
                self.save_model()
                print(f"  ✓ Best model saved (AUC: {val_auc:.4f})\n")

    def save_model(self):
        """Save model checkpoint"""
        Path(self.config.model_path).parent.mkdir(exist_ok=True, parents=True)

        checkpoint = {
            'model_state_dict': self.model.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
            'config': self.config,
            'history': self.history
        }
        torch.save(checkpoint, self.config.model_path)

    def plot_training_curves(self):
        """Visualize training progress"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

        epochs = range(1, len(self.history['train_loss']) + 1)

        # Loss curves
        ax1.plot(epochs, self.history['train_loss'], 'b-', label='Train Loss', linewidth=2)
        ax1.plot(epochs, self.history['val_loss'], 'r-', label='Val Loss', linewidth=2)
        ax1.set_xlabel('Epoch', fontsize=12)
        ax1.set_ylabel('Loss', fontsize=12)
        ax1.set_title('Training & Validation Loss', fontsize=14, fontweight='bold')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # AUC curves
        ax2.plot(epochs, self.history['train_auc'], 'b-', label='Train AUC', linewidth=2)
        ax2.plot(epochs, self.history['val_auc'], 'r-', label='Val AUC', linewidth=2)
        ax2.set_xlabel('Epoch', fontsize=12)
        ax2.set_ylabel('AUC-ROC', fontsize=12)
        ax2.set_title('Training & Validation AUC', fontsize=14, fontweight='bold')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        ax2.set_ylim([0.5, 1.0])

        plt.tight_layout()
        plt.savefig(f"{self.config.results_path}/training_curves.png", dpi=300)
        print(f"✓ Training curves saved")


def main():
    """Main pipeline"""
    print("\n" + "="*70)
    print("  HIGGS BOSON DETECTION with PyTorch")
    print("  Using ATLAS CSV Dataset")
    print("="*70 + "\n")

    # Configuration
    config = Config()

    # Load CSV dataset
    print("[1/5] Loading ATLAS CSV data...")
    full_dataset = ATLASDataset(config.data_path)

    # Auto-detect input features
    config.input_features = full_dataset.features.shape[1]
    print(f"  Auto-detected input features: {config.input_features}")

    # Split into train/val
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(
        full_dataset, [train_size, val_size]
    )

    print(f"  Train: {train_size:,} events")
    print(f"  Val: {val_size:,} events\n")

    # Data loaders
    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False, num_workers=0)

    # Train model
    print("[2/5] Training Higgs classifier...")
    trainer = DistributedTrainer(config)
    trainer.fit(train_loader, val_loader)

    # Plot results
    print("\n[3/5] Generating visualizations...")
    trainer.plot_training_curves()

    print("\n" + "="*70)
    print("✓ TRAINING COMPLETE")
    print("="*70)
    print(f"\nModel saved to: {config.model_path}")
    print(f"Results saved to: {config.results_path}")
    print("\n")


if __name__ == "__main__":
    main()
```
